```python
### Check and combine countries for similiarities. Wait for Results. It may take a minute.

import pandas as pd
import os
import glob
from datetime import datetime

# Versionsnummer mit Datum und Uhrzeit
version = datetime.now().strftime("# Version: %Y-%m-%d %H:%M:%S")
print(version)

# Verzeichnis definieren
path = r"01_CSV_Ressources\*.csv"

# Liste aller CSV-Dateien im Verzeichnis mit glob
all_files = glob.glob(path)

# Eine Masterliste aller Länder erstellen
all_countries = set()

# Für jede Datei die Daten in einem großen DataFrame zusammenführen
country_check_df = pd.DataFrame({'Entity': list(all_countries)})
country_check_df.set_index('Entity', inplace=True)

for file in all_files:
    try:
        df = pd.read_csv(file, on_bad_lines='skip')
        if 'Entity' in df.columns:
            # Dateiname oder Nummer als Spaltenname verwenden
            column_name = os.path.basename(file)
            df = df[['Entity']].drop_duplicates()
            df.set_index('Entity', inplace=True)
            df[column_name] = 1  # matched
            country_check_df = country_check_df.join(df, how="outer")
    except Exception as e:
        print(f"Error reading file {file}. Error: {e}")

country_check_df.fillna(0, inplace=True)
country_check_df = country_check_df.astype(int)  # Ganzzahlen

## Delete wrong entities
for name in excluding_list:
    to_delete = country_check_df[country_check_df.index.str.contains(name, case=False, na=False, regex=False)].index
    country_check_df.drop(to_delete, inplace=True)

# Create a new dataframe for our combined data
master_df = pd.DataFrame()

remaining_countries = country_check_df.index.tolist()  # Get the remaining countries after filtering

for file in all_files:
    try:
        df = pd.read_csv(file, on_bad_lines='skip')
        if 'Entity' in df.columns and 'Year' in df.columns:
            df = df[df['Entity'].isin(remaining_countries)]  # Filter by the remaining countries
            df = df[(df['Year'] >= 1965) & (df['Year'] <= 2022)]
            df.fillna({"Entity": "", "Code": "", df.columns[-1]: 0}, inplace=True)
            df.set_index(['Entity', 'Year'], inplace=True)
            if master_df.empty:
                master_df = df
            else:
                master_df = master_df.join(df[df.columns[-1]], how="outer")
    except Exception as e:
        print(f"Error processing file {file}. Error: {e}")

# Resetting index for final output
master_df.reset_index(inplace=True)

#print("Master DataFrame:\n", master_df)

# Erstellen eines ExcelWriter-Objekts
output_path = '09_Outputs/countryclean.xlsx'
with pd.ExcelWriter(output_path) as writer:
    matches_per_file.to_excel(writer, sheet_name='Übereinstimmungen pro Datei', index=False)
    files_per_country.to_excel(writer, sheet_name='Dateien pro Land', index=True)
    country_check_df.to_excel(writer, sheet_name='Country Check DataFrame', index=True)
    master_df.to_excel(writer, sheet_name='Combined Data 1965-2022', index=False)
```
