```python
## Only run the script if something went wrong with setting the quotes for the entity column

import os
import pandas as pd
import glob

# Verzeichnis definieren
path = r"C:\Users\Alpha\OneDrive - Seolotsen\03_SEO\01_1_Data Analyst Portfolio\Git_Repo\Stackfuel_Data_Analyst\01_CSV_Ressources\*.csv"

# Liste aller CSV-Dateien im Verzeichnis mit glob
all_files = glob.glob(path)

# Listen zum Speichern der Dateinamen
successful_files = []
error_files = []

for file in all_files:
    try:
        # Zeilenweise Datei einlesen und bearbeiten
        with open(file, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            new_lines = [line.replace('"', '') for line in lines]

        # Geänderte Zeilen zurück in Datei schreiben
        with open(file, 'w', encoding='utf-8') as f:
            f.writelines(new_lines)

        successful_files.append(file)
    except Exception as e:
        print(f"Fehler beim Bearbeiten der Datei {file}: {e}")
        error_files.append(file)

print("\nErfolgreich bearbeitete Dateien:")
for f in successful_files:
    print(f)

print("\nDateien mit Fehlern:")
for f in error_files:
    print(f)
```

```python
## Setting the quotation marks for the entity column. Only hit one times,
## otherwise, you have 2,3 quotation mark

import os
import pandas as pd
import glob

# Verzeichnis definieren
path = r"C:\Users\Alpha\OneDrive - Seolotsen\03_SEO\01_1_Data Analyst Portfolio\Git_Repo\Stackfuel_Data_Analyst\01_CSV_Ressources\*.csv"

# Liste aller CSV-Dateien im Verzeichnis mit glob
all_files = glob.glob(path)

# Listen zum Speichern der Dateinamen
successful_files = []
error_files = []

for file in all_files:
    try:
        # Automatisches Erkennen des richtigen Trennzeichens (Komma oder Semikolon)
        with open(file, 'r', encoding='utf-8') as f:
            first_line = f.readline()
            delimiter = ',' if ',' in first_line else ';'
        
        # CSV-Datei einlesen
        df = pd.read_csv(file, delimiter=delimiter)
        
        # Anwenden von Anführungszeichen nur auf die "Entity" Spalte
        if 'Entity' in df.columns:
            df['Entity'] = df['Entity'].apply(lambda x: f'"{x}"' if '"' not in x else x)

        # Die geänderte CSV-Datei speichern
        df.to_csv(file, index=False, sep=',')

        successful_files.append(file)
    except Exception as e:
        print(f"Fehler beim Bearbeiten der Datei {file}: {e}")
        error_files.append(file)

print("\nErfolgreich bearbeitete Dateien:")
for f in successful_files:
    print(f)

print("\nDateien mit Fehlern:")
for f in error_files:
    print(f)
```

```python

import time

# Verzeichnispfad definieren, in dem sich Ihre CSV-Dateien befinden
path = r"C:\Users\Alpha\OneDrive - Seolotsen\03_SEO\01_1_Data Analyst Portfolio\Git_Repo\Stackfuel_Data_Analyst\01_CSV_Ressources"

# Alle Dateien im Verzeichnis durchlaufen
for filename in os.listdir(path):
    # Überprüfen, ob die Datei eine CSV-Datei ist
    if filename.endswith(".csv"):
        full_path = os.path.join(path, filename)
        
        # Aktuellen Zeitstempel holen
        current_time = time.time()
        
        # Änderungsdatum und Zugriffsdatum der Datei auf den aktuellen Zeitstempel setzen
        os.utime(full_path, (current_time, current_time))

print("Zeitstempel aller CSV-Dateien aktualisiert!")

```

```python
## Show the amount of countries
# Verzeichnis definieren
path = r"C:\Users\Alpha\OneDrive - Seolotsen\03_SEO\01_1_Data Analyst Portfolio\Git_Repo\Stackfuel_Data_Analyst\01_CSV_Ressources\*.csv"

# Liste aller CSV-Dateien im Verzeichnis mit glob
all_files = glob.glob(path)

# Listen zum Speichern der Dateinamen und der Anzahl der Länder
file_names = []
countries_count = []

for file in all_files:
    try:
        # CSV-Datei einlesen
        df = pd.read_csv(file, on_bad_lines='skip')
        
        # Überprüfen, ob "Entity" in den Spaltennamen ist
        if 'Entity' in df.columns:
            # Anzahl der eindeutigen Länder in der "Entity" Spalte zählen
            unique_countries = df['Entity'].nunique()
            file_names.append(os.path.basename(file))
            countries_count.append(unique_countries)
    except Exception as e:
        print(f"Error reading file {file}. Error: {e}")

# Daten in einem DataFrame speichern
result_df = pd.DataFrame({
    'File Name': file_names,
    'Unique Countries': countries_count
})

print(result_df)
```
